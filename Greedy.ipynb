## Greedy

```
- 현재 상황에서 지금 당장 좋은 것만 고르는 방법
- 기준에 따라 좋은 것을 선택하는 알고리즘이므로, '가장 큰 순서대로', '가장 작은 순서대로'와 같은 기준을 알게 모르게 제시해준다.
- 대체로 이 기준은 정렬 알고리즘을 사용했을 때 만족시킬 수 있으므로 그리디 알고리즘 문제는 자주 정렬 알고리즘과 짝을 이뤄 출제된다.
```

````python
#@title 예제3-1 거스름돈

# 문제 내용 생략
# 예시) N = 1260 일때 500원 2개, 100원 2개, 50원 1개, 10원 1개

# 방향
# // 몫 % 나머지 사용
# 각 화폐별로 나누고 몫과 나머지를 구하고 남은 돈 업데이트하는 식으로 풀이

# 코드
n = 1260
count = 0

# 큰 단위의 화폐부터 차례대로 확인
coin_types = [500, 100, 50, 10]

for coin in coin_types:
    count += n // coin # 해당 화폐로 거슬러 줄 수 있는 동전 개수
    n %= coin

print(count)
````

````python
# 각 동전의 갯수 출력
n = int(input())
count = 0

# 동전 개수 변수 추가
cnt_500 = 0
cnt_100 = 0
cnt_50 = 0
cnt_10 = 0

# 큰 단위의 화폐부터 차례대로 확인
coin_types = [500, 100, 50, 10]

for coin in coin_types:
    if n // coin > 0:  # 해당 화폐로 거슬러 줄 수 있는 동전이 있을 때
        coin_count = n // coin  # 해당 화폐로 거슬러 줄 수 있는 동전 개수
        count += coin_count  # 동전 개수 총합에 더하기
        n %= coin  # 나머지 금액 업데이트

        # 각 동전 종류별 개수 저장
        if coin == 500:
            cnt_500 = coin_count
        elif coin == 100:
            cnt_100 = coin_count
        elif coin == 50:
            cnt_50 = coin_count
        elif coin == 10:
            cnt_10 = coin_count

# 결과 출력
print(f"500원: {cnt_500}개, 100원: {cnt_100}개, 50원: {cnt_50}개, 10원: {cnt_10}개, 총 동전 개수: {count}")
````
